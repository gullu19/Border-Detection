```python
import cv2
import rembg
import numpy as np

# Variable to specify the input image path
image_path = "C:/Users/Admin/Documents/AI Engineer - ProductizeTech Assignment/TEST IMAGES/1.jpg"

# Read and display the image
image = cv2.imread(image_path)
image = cv2.resize(image, (1200,700))

# Create a loop to allow the user to perform the desired actions multiple times
while True:
    # Ask the user to draw a rectangle over an object inside the image
    roi = cv2.selectROI("Select ROI", image)
    cv2.destroyAllWindows()

    # Crop the selected Region Of Interest (ROI)
    x, y, w, h = roi
    cropped_image = image[y:y+h, x:x+w]

    # Remove the background from the cropped image using rembg
    output_image = rembg.remove(cropped_image)

    # Split the output image into RGB channels and the alpha channel
    output_bgr = output_image[:, :, :3]
    alpha_mask = output_image[:, :, 3] / 255.0  # Normalize alpha values to range 0-1

    # Create a mask from the alpha channel
    alpha_mask = np.repeat(alpha_mask[:, :, np.newaxis], 3, axis=2)

    # Overlay the object outline on the original input image using the alpha mask
    blended_image = (alpha_mask * output_bgr) + ((1 - alpha_mask) * cropped_image)

    # Convert the blended image back to uint8 format
    blended_image = blended_image.astype(np.uint8)

    # Draw an outline over the object
    contours, _ = cv2.findContours(cv2.cvtColor(output_bgr, cv2.COLOR_BGR2GRAY),
                                   cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(blended_image, contours, -1, (0, 255, 0), 2)

    # Overlay the object outline in the original input image
    image[y:y+h, x:x+w] = blended_image

    # Show the output image
    cv2.imshow("Output Image", image)

    # Wait for key press
    key = cv2.waitKey(0)

    # Check if the user wants to quit or clear the outline
    if key == ord("q"):
        break
    elif key == ord("c"):
        image = cv2.imread(image_path)

# Clean up
cv2.destroyAllWindows()
```
